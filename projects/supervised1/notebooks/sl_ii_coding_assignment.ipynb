```python
from matplotlib import pyplot as plt
import numpy as np

X, y = mnist["data"], mnist["target"]
print(X.shape)
for i in range(4):
    _, axs = plt.subplots(nrows=1, ncols=4)
    for j in range(4):
        s = np.random.randint(0, X.shape[0])
        axs[j].imshow(X[s].reshape(28, 28))
        axs[j].set_title(y[s])
```
